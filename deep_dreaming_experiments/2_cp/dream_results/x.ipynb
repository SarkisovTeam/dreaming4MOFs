```python
import pandas as pd
```

```python
df = pd.read_csv('./cp_min_1000_dream_results_DONE.csv')
```

```python
import uuid

def generate_truncated_uuids(n, length=8):
    ids = []
    for _ in range(n):
        unique_id = str(uuid.uuid4()).replace('-', '')[:length]
        ids.append(unique_id)
    return ids

# Generate N unique identifiers
N = len(df)
unique_ids = generate_truncated_uuids(N, length=8)
df['linker_hash'] = unique_ids
```

```python
df.head(1)
```

```python
dreamed_mof_name = []

for iter,row in df.iterrows():
    parts = row['seed_mof_name'].split('_')
    dreamed_mof_name.append(f'{parts[0]}_{parts[1]}_{row["linker_hash"]}')
    print(f'{parts[0]}_{parts[1]}_{row["linker_hash"]}')

df['dreamed_mof_name'] = dreamed_mof_name
```

```python
df.to_csv('./cp_min_1000_dream_results_DONE.csv', index=False)
```
